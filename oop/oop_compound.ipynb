```python
from math import pow
from typing import Union, List
```

```python
class CompoundInterest:
    def __init__(self, year:int, interest:float, sum:float):
        self.year = year
        self.interest = interest
        self.sum = sum

    def __lt__(self, other):
        return self.sum < other.sum
```

```python
def compound_interest(start_sum: Union[int, float],
                      interest: Union[int, float],
                      no_of_year: int,
                      no_compute: int = 1) -> float:
    return start_sum * pow((1 + interest / no_compute), no_compute * no_of_year)
```

```python
start_sum = 400_000  # Målsumman
interest = 0.05  # Årlig ränta (5%)
no_compute = 1  # Antal gånger räntan beräknas per år (1 gång)
no_of_year = 10  # Antal år

serie = [0.03, 0.04, 0.05, 0.06, 0.07]

all = []
```

```python
for no_of_year in range(3, 8):
    # print(f"År:{t}")
    for interest in serie:
        P = compound_interest(start_sum=start_sum, interest=interest, no_of_year=no_of_year)
        all.append(CompoundInterest(year=no_of_year, interest=round(interest * 100), sum=round(P)))
```

```python
# Alternatively, you could sort the list in place:
all.sort(reverse=True)
```

```python
# Print sorted list
for compound_interest in all:
    print(f"Year: {compound_interest.year}, Ranta: {compound_interest.interest}%, Sum: {compound_interest.sum}")
```
